```
samples = 1280
lines = 1024
bands = 151
print('Expected size:', samples*bands*lines)
print('Actual szie:', raw_image.shape)

spectral_image = numpy.reshape(raw_image, (lines, bands, samples))

plt.imshow(spectral_image[:,40,:], cmap="gray")
plt.show()
#plt.plot(spectral_image[500,:,500])
#plt.show()
```

```
# Select white sample 

y1 = 410
y2 = 440
x1 = 80
x2 = 100

preview_selected = spectral_image[:,100,:].copy()
preview_selected[x1:x2,y1:y2] = 250

plt.imshow(preview_selected, cmap="gray")
plt.show()
```

```
# get average for white
white_image = spectral_image[x1:x2,:,y1:y2]
white_av1 = numpy.sum(white_image, axis = 0) # get average for White
white_av = numpy.sum(white_av1, axis = 1)/((x2-x1)*(y2-y1)) # get average for White
print(white_av.shape)
for i in range(1,10):
  plt.plot(white_image[10,:,10], 'b')
plt.plot(white_av, 'r')
plt.show()
```

```
# Correct spectral_image
corrected_image = numpy.zeros((lines, bands, samples), dtype=numpy.float32)
for i in range(lines):
  for j in range(samples):
      corrected_image[i,:,j] = numpy.divide(spectral_image[i,:,j], white_av)

plt.rcParams['figure.dpi'] = 200
plt.imshow(corrected_image[:,100,:], cmap="gray")
plt.show()
plt.rcParams['figure.dpi'] = 200
plt.plot(100*corrected_image[500,:,400])
plt.plot(100*corrected_image[80,:,410], 'r')
plt.plot(100*corrected_image[100,:,200], 'r')
#plt.plot(white_av)

plt.yticks(range(0, 130, 10))
#plt.xticks(range(200, 2500, 200))
#plt.grid(b=True, linestyle='-')
#plt.xlabel('xlabel', fontsize=10)
#plt.ylabel('ylabel', fontsize=10)
#plt.tick_params(axis='both', which='major', labelsize=10)
plt.xlabel('Bands')
plt.ylabel('Reflectance [0-100]%')

plt.show()

```
