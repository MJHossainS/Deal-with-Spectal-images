```python
from os import listdir                 # read the dir conent
from os.path import isfile, join # working with paths in operating system
import pandas as pd                   # pandas to read the csv files
import matplotlib.pyplot as plt 
```

```python
directory='/content/drive/MyDrive/ASI/Copy of LectureExercise #1, CSV/csv files'
#directory = sorted(directory)
filename1='01 paper green.Sample.Raw.csv'  #  paper green
filename2='24 carton white.Sample.Raw.csv' 
filename3='30 plastic white.Sample.Raw.csv' 
filename4='33 candle parafin green.Sample.Raw.csv' 
filename5='50 textile towel green.Sample.Raw.csv' 
filename6='53 wool yellow.Sample.Raw.csv' 
filename7='56 ink orange.Sample.Raw.csv' 
filename8='71 live leaf ivy.Sample.Raw.csv' 
```

```python
csv_file1 = join(directory, filename1) # generate path to csv file
data1 = pd.read_csv(csv_file1)            
waves1 = data1['nm']                           # get waves array
reflectance1 = data1[' %R']                  # get reflectance array

csv_file2 = join(directory, filename2) # generate path to csv file
data2 = pd.read_csv(csv_file2)            
waves2 = data2['nm']                           # get waves array
reflectance2 = data2[' %R']                  # get reflectance array

csv_file3 = join(directory, filename3) # generate path to csv file
data3 = pd.read_csv(csv_file3)            
waves3 = data3['nm']                           # get waves array
reflectance3 = data3[' %R']                  # get reflectance array

csv_file4 = join(directory, filename4) # generate path to csv file
data4 = pd.read_csv(csv_file4)            
waves4 = data4['nm']                           # get waves array
reflectance4 = data4[' %R']                  # get reflectance array

csv_file5 = join(directory, filename5) # generate path to csv file
data5 = pd.read_csv(csv_file5)            
waves5 = data5['nm']                           # get waves array
reflectance5 = data5[' %R']                  # get reflectance array

csv_file6 = join(directory, filename6) # generate path to csv file
data6 = pd.read_csv(csv_file6)            
waves6 = data6['nm']                           # get waves array
reflectance6 = data6[' %R']                  # get reflectance array

csv_file7 = join(directory, filename7) # generate path to csv file
data7 = pd.read_csv(csv_file7)            
waves7 = data7['nm']                           # get waves array
reflectance7 = data7[' %R']                  # get reflectance array

csv_file8 = join(directory, filename8) # generate path to csv file
data8 = pd.read_csv(csv_file8)            
waves8 = data8['nm']                           # get waves array
reflectance8 = data8[' %R']                  # get reflectance array
```

```python
#Plot single spectrum
f = plt.figure()
f.set_figwidth(16)
f.set_figheight(4)

plt. plot(waves1, reflectance1, 'r', label='paper green')  # plot spectrum in red color
plt. plot(waves2, reflectance2, 'b', label='carton white')  # plot spectrum in blue  color
plt. plot(waves3, reflectance3, 'g', label='plastic white')
plt. plot(waves4, reflectance4, 'c', label='candle parafin green')
plt. plot(waves5, reflectance5, 'm', label='textile towel green')
plt. plot(waves6, reflectance6, 'y', label='wool yellow')
plt. plot(waves7, reflectance7, 'k', label='ink orange')
plt. plot(waves8, reflectance8, 'brown', label='live leaf ivy')



plt.yticks(range(1, 110, 10))
plt.xticks(range(200, 2500, 200))
plt.grid(b=True, linestyle='-')
plt.xlabel('xlabel', fontsize=12)
plt.ylabel('ylabel', fontsize=12)
plt.tick_params(axis='both', which='major', labelsize=8)
plt.xlabel('Wavelength, nm')
plt.ylabel('Reflectance [0-100]%')
plt.legend()
plt.show()
```
